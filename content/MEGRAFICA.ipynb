```python
from IPython.display import IFrame
import base64

html = """<!DOCTYPE html>
<html><head>
<meta charset="UTF-8">
<style>
*{box-sizing:border-box;margin:0;padding:0}
body{background:#0a1628;font-family:monospace;overflow:hidden}
#wrap{display:flex;width:1000px;height:580px}
#left{width:155px;background:#0d1f3c;border-right:1px solid #1a3a5c;
  padding:8px;display:flex;flex-direction:column;gap:6px;flex-shrink:0}
.pt{color:#00aaff;font-size:7px;text-align:center;border-bottom:1px solid #1a3a5c;padding-bottom:4px;margin-bottom:2px}
.sr{display:flex;flex-direction:column;gap:2px}
.sr label{font-size:7px;color:#4499cc;display:flex;justify-content:space-between}
.sr label span{color:#00ddff;font-weight:bold}
input[type=range]{width:100%;-webkit-appearance:none;height:3px;background:#1a3a5c;border-radius:2px;outline:none}
input[type=range]::-webkit-slider-thumb{-webkit-appearance:none;width:10px;height:10px;border-radius:50%;background:#00aaff;cursor:pointer}
#right{flex:1;position:relative;overflow:hidden}
#cc{}
#toggles{position:absolute;top:8px;right:8px;display:flex;flex-direction:column;gap:5px;z-index:10}
.tr{display:flex;align-items:center;gap:6px;background:rgba(0,20,40,0.75);
  border:1px solid #1a3a5c;border-radius:20px;padding:4px 10px;cursor:pointer;user-select:none}
.ts{width:28px;height:14px;background:#1a3a5c;border-radius:7px;position:relative;transition:background .2s}
.ts::after{content:'';position:absolute;width:10px;height:10px;background:#555;
  border-radius:50%;top:2px;left:2px;transition:all .2s}
.ts.on{background:#005588}
.ts.on::after{left:16px;background:#00aaff}
.tl{font-size:8px;color:#4499cc;white-space:nowrap}
#btns{position:absolute;bottom:10px;right:10px;display:flex;gap:6px;z-index:10}
.cb{width:42px;height:42px;border-radius:50%;background:radial-gradient(circle at 35% 35%,#1a4a7a,#071428);
  border:1px solid #2a6a9a;color:#00aaff;font-size:7px;cursor:pointer;
  display:flex;flex-direction:column;align-items:center;justify-content:center;gap:1px;user-select:none}
.cb:hover{border-color:#00aaff}
.cb span{font-size:13px;line-height:1}
.cb.stop{background:radial-gradient(circle at 35% 35%,#3a0a0a,#1a0505);border-color:#f44;color:#f44}
#dc{display:none}
</style>
</head><body>
<div id="wrap">
  <div id="left">
    <div class="pt">PARAMETERS</div>
    <div class="sr"><label>Vision <span id="vv">60</span></label>
      <input type="range" id="sv" min="20" max="150" value="60"
        oninput="P.vision=+this.value;document.getElementById('vv').textContent=this.value"></div>
    <div class="sr"><label>Cohesion <span id="cv">0.8</span></label>
      <input type="range" id="sc" min="0" max="2" step="0.1" value="0.8"
        oninput="P.cohesion=+this.value;document.getElementById('cv').textContent=(+this.value).toFixed(1)"></div>
    <div class="sr"><label>Separacion <span id="sepv">1.2</span></label>
      <input type="range" id="ss" min="0" max="3" step="0.1" value="1.2"
        oninput="P.separation=+this.value;document.getElementById('sepv').textContent=(+this.value).toFixed(1)"></div>
    <div class="sr"><label>Alineacion <span id="av">1.0</span></label>
      <input type="range" id="sa" min="0" max="2" step="0.1" value="1.0"
        oninput="P.alignment=+this.value;document.getElementById('av').textContent=(+this.value).toFixed(1)"></div>
    <div class="sr"><label>Velocidad <span id="spv">2.5</span></label>
      <input type="range" id="ssp" min="1" max="6" step="0.1" value="2.5"
        oninput="P.maxSpeed=+this.value;document.getElementById('spv').textContent=(+this.value).toFixed(1)"></div>
    <div class="sr"><label>Peces <span id="fv">40</span></label>
      <input type="range" id="sf" min="10" max="100" value="40"
        oninput="setFishCount(+this.value)"></div>
  </div>
  <div id="right">
    <div id="cc"></div>
    <div id="toggles">
      <div class="tr" onclick="toggleHunger()">
        <div class="ts" id="ht"></div><span class="tl" id="hl">Sin Hambre</span></div>
      <div class="tr" id="dc" onclick="toggleDeath()">
        <div class="ts on" id="dt"></div><span class="tl" id="dl">Con Muerte</span></div>
      <div class="tr" onclick="toggleAuto()">
        <div class="ts" id="at"></div><span class="tl" id="al">Auto OFF</span></div>
      <div class="tr" onclick="toggleWall()">
        <div class="ts on" id="wt"></div><span class="tl" id="wl">Rebote</span></div>
    </div>
    <div id="btns">
      <div class="cb stop" onclick="stopSim()"><span>⏹</span>Stop</div>
      <div class="cb" onclick="randomize()"><span>🎲</span>Random</div>
      <div class="cb" onclick="resetSim()"><span>↻</span>Reset</div>
    </div>
  </div>
</div>

<script>
// ── Estado global ─────────────────────────────────────────────
var P={vision:60,cohesion:0.8,separation:1.2,alignment:1.0,maxSpeed:2.5,fishCount:40};
var hungerOn=false,deathOn=true,autoOn=false,wallOn=true;
var fishes=[],pellets=[];
var CW=840,CH=580;

// ── Toggles ───────────────────────────────────────────────────
function toggleHunger(){
  hungerOn=!hungerOn;
  document.getElementById('ht').classList.toggle('on',hungerOn);
  document.getElementById('hl').textContent=hungerOn?'Con Hambre':'Sin Hambre';
  document.getElementById('dc').style.display=hungerOn?'flex':'none';
  if(!hungerOn) for(var f of fishes){f.hunger=0;f.blinkState=false;}
}
function toggleDeath(){
  deathOn=!deathOn;
  document.getElementById('dt').classList.toggle('on',deathOn);
  document.getElementById('dl').textContent=deathOn?'Con Muerte':'Sin Muerte';
}
function toggleAuto(){
  autoOn=!autoOn;
  document.getElementById('at').classList.toggle('on',autoOn);
  document.getElementById('al').textContent=autoOn?'Auto ON':'Auto OFF';
}
function toggleWall(){
  wallOn=!wallOn;
  document.getElementById('wt').classList.toggle('on',wallOn);
  document.getElementById('wl').textContent=wallOn?'Rebote':'Atravesar';
}
function setFishCount(n){
  document.getElementById('fv').textContent=n;
  P.fishCount=n;
  while(fishes.length<n) fishes.push(new Fish());
  while(fishes.length>n) fishes.pop();
}
function stopSim(){
  if(window._p5i){window._p5i.noLoop();window._p5i.remove();}
  document.getElementById('cc').innerHTML=
    '<div style="color:#f44;font-family:monospace;padding:30px">⏹ Detenido</div>';
}
function resetSim(){fishes=[];for(var i=0;i<P.fishCount;i++)fishes.push(new Fish());}
function randomize(){
  P.vision=Math.floor(Math.random()*130+20);
  P.cohesion=+(Math.random()*2).toFixed(1);
  P.separation=+(Math.random()*3).toFixed(1);
  P.alignment=+(Math.random()*2).toFixed(1);
  P.maxSpeed=+(Math.random()*5+1).toFixed(1);
  document.getElementById('sv').value=P.vision; document.getElementById('vv').textContent=P.vision;
  document.getElementById('sc').value=P.cohesion; document.getElementById('cv').textContent=P.cohesion;
  document.getElementById('ss').value=P.separation; document.getElementById('sepv').textContent=P.separation;
  document.getElementById('sa').value=P.alignment; document.getElementById('av').textContent=P.alignment;
  document.getElementById('ssp').value=P.maxSpeed; document.getElementById('spv').textContent=P.maxSpeed;
}

// ── Cargar p5.js dinámicamente ────────────────────────────────
var _sc=document.createElement('script');
_sc.src='https://cdnjs.cloudflare.com/ajax/libs/p5.js/1.9.0/p5.min.js';
_sc.onload=function(){

  // Datos oceánicos
  var OD={temperature:18.5,pressure:1013.2,windSpeed:12.3,
          windDirection:0,waveHeight:1.8,salinity:35.2,
          radarSweep:0,sonarPings:[]};

  window._p5i = new p5(function(p){

    p.setup=function(){
      var cv=p.createCanvas(CW,CH); cv.parent('cc');
      cv.mousePressed(function(){
        if(p.mouseX<0||p.mouseX>CW||p.mouseY<0||p.mouseY>CH) return;
        if(p.mouseX>CW-200&&p.mouseY<180) return;
        if(p.mouseX>CW-100&&p.mouseY>CH-180) return;
        var n=p.floor(p.random(3,6));
        for(var i=0;i<n;i++) pellets.push(new Pellet(p,p.mouseX+p.random(-15,15),p.mouseY+p.random(-15,15)));
      });
      for(var i=0;i<P.fishCount;i++) fishes.push(new Fish(p));
      Fish.prototype._p=p;
      Pellet.prototype._p=p;
    };

    p.draw=function(){
      // Agua
      for(var y=0;y<CH;y++){
        var c=p.lerpColor(p.color(100,180,230),p.color(40,120,180),y/CH);
        p.stroke(c);p.line(0,y,CW,y);
      }
      // Titulo
      p.noStroke();p.textAlign(p.LEFT,p.TOP);
      p.textStyle(p.BOLD);p.textSize(22);p.textFont('Georgia,serif');
      p.fill(0,0,0,120);p.text("Welcome to Juanito's Reef",22,22);
      p.fill(20,60,120);
      for(var dx=-1;dx<=1;dx++) for(var dy=-1;dy<=1;dy++) if(dx||dy) p.text("Welcome to Juanito's Reef",20+dx,20+dy);
      p.fill(255);p.text("Welcome to Juanito's Reef",20,20);
      p.textSize(10);p.textStyle(p.NORMAL);p.textFont('monospace');
      p.fill(255,255,255,200);p.text('~ Interactive Boids Simulation ~',22,48);

      // Auto pellets
      if(autoOn&&p.frameCount%60===0){
        var ax=p.random(50,CW-50),ay=p.random(20,100);
        for(var i=0;i<p.floor(p.random(2,5));i++) pellets.push(new Pellet(p,ax+p.random(-10,10),ay));
      }

      // Pellets
      for(var i=pellets.length-1;i>=0;i--){
        pellets[i].update();pellets[i].show();
        if(pellets[i].consumed||pellets[i].pos.y>CH-10) pellets.splice(i,1);
      }

      // Fish
      for(var i=fishes.length-1;i>=0;i--){
        fishes[i].flock(fishes);
        if(hungerOn) fishes[i].seekFood(pellets);
        fishes[i].update();fishes[i].edges();fishes[i].show();
        if(fishes[i].dead&&fishes[i].opacity<=0){
          fishes.splice(i,1);P.fishCount--;
          document.getElementById('sf').value=P.fishCount;
          document.getElementById('fv').textContent=P.fishCount;
        }
      }

      // Naval displays
      updateOcean(p,OD);
      drawNaval(p,OD);
    };

    function updateOcean(p,od){
      if(p.frameCount%10===0){od.temperature+=p.random(-0.02,0.02);od.temperature=p.constrain(od.temperature,15,22);}
      if(p.frameCount%15===0){od.pressure+=p.random(-0.1,0.1);od.pressure=p.constrain(od.pressure,1008,1018);}
      if(p.frameCount%12===0){od.windSpeed+=p.random(-0.15,0.15);od.windSpeed=p.constrain(od.windSpeed,8,20);}
      od.windDirection=(od.windDirection+p.random(-1,1)+360)%360;
      if(p.frameCount%20===0){od.waveHeight+=p.random(-0.02,0.02);od.waveHeight=p.constrain(od.waveHeight,0.5,3.0);}
      if(p.frameCount%25===0){od.salinity+=p.random(-0.005,0.005);od.salinity=p.constrain(od.salinity,34.5,36.0);}
      od.radarSweep=(od.radarSweep+0.9)%360;
      if(p.frameCount%30===0&&od.sonarPings.length<5)
        od.sonarPings.push({angle:p.random(360),distance:p.random(30,80),life:60});
      for(var i=od.sonarPings.length-1;i>=0;i--){od.sonarPings[i].life--;if(od.sonarPings[i].life<=0)od.sonarPings.splice(i,1);}
    }

    function drawNaval(p,od){
      p.push();
      var px=15,py=CH-95,pw=140,ph=82;
      p.fill(0,20,30,180);p.stroke(0,180,200,150);p.strokeWeight(1);p.rect(px,py,pw,ph,5);
      p.stroke(0,200,220,200);p.line(px,py+8,px+pw,py+8);
      p.noStroke();p.fill(0,220,255);p.textSize(7);p.textFont('monospace');p.textAlign(p.LEFT,p.TOP);
      p.text('OCEAN SURVEILLANCE',px+4,py+2);
      // Radar
      p.push();p.translate(px+35,py+40);
      p.fill(0,10,15,200);p.stroke(0,150,170,100);p.strokeWeight(0.5);p.circle(0,0,54);
      p.noFill();p.stroke(0,180,200,80);
      for(var i=1;i<=3;i++) p.circle(0,0,54*i/3);
      p.stroke(0,180,200,60);p.line(-27,0,27,0);p.line(0,-27,0,27);
      var sa=p.radians(od.radarSweep);
      p.stroke(0,255,200,200);p.strokeWeight(1.5);p.line(0,0,p.cos(sa)*27,p.sin(sa)*27);
      for(var i=0;i<40;i++){var a=sa-p.radians(i*3);p.stroke(0,255,200,p.map(i,0,40,80,0));p.strokeWeight(0.5);p.line(0,0,p.cos(a)*27,p.sin(a)*27);}
      for(var i=0;i<od.sonarPings.length;i++){
        var pg2=od.sonarPings[i],pa=p.radians(pg2.angle),pd=p.map(pg2.distance,0,100,0,27),al=p.map(pg2.life,0,60,0,255);
        p.fill(0,255,150,al);p.noStroke();p.circle(p.cos(pa)*pd,p.sin(pa)*pd,2);
      }
      p.fill(0,220,255);p.noStroke();p.textSize(6);p.textAlign(p.CENTER,p.TOP);p.text('SONAR',0,30);
      p.pop();
      // Datos
      p.noStroke();p.fill(0,220,255);p.textSize(6);p.textFont('monospace');p.textAlign(p.LEFT,p.TOP);
      var lh=7,dx2=px+72,dy2=py+15;
      p.text('TEMP: '+od.temperature.toFixed(1)+'C',dx2,dy2);
      p.text('PRES: '+od.pressure.toFixed(0),dx2,dy2+lh);
      p.text('WIND: '+od.windSpeed.toFixed(1),dx2,dy2+lh*2);
      p.text('SAL:  '+od.salinity.toFixed(1),dx2,dy2+lh*3);
      // Wave graph
      var gx=px+72,gy=py+50,gw=60,gh=20;
      p.fill(0,10,15,150);p.stroke(0,180,200,100);p.strokeWeight(0.5);p.rect(gx,gy,gw,gh,2);
      var ts2=p.map(P.maxSpeed,1,6,0.3,1.5);
      p.noFill();p.stroke(0,255,200,200);p.strokeWeight(1);p.beginShape();
      for(var i=0;i<gw;i+=2){
        var ph2=(p.frameCount*ts2*0.1)+(i*0.15);
        var wv=(p.sin(ph2)*0.5+p.sin(ph2*2.3+p.sin(ph2*0.3)*2)*0.3+p.sin(ph2*4.7+p.cos(ph2*0.7)*3)*0.2)*
               (p.sin(ph2*0.2)*0.5+0.5)*0.8*od.waveHeight*2.5;
        p.vertex(gx+i,gy+gh/2+wv);
      }
      p.endShape();
      p.fill(0,220,255);p.noStroke();p.textSize(5);p.textAlign(p.LEFT,p.TOP);
      p.text('ACOUSTIC:'+od.waveHeight.toFixed(1)+'m',gx+2,gy+2);
      // Heat map
      var mx=px+pw+10,my=py,mw=70,mh=82,gs=10;
      var cols=Math.floor(mw/gs),rows=Math.floor(mh/gs)-1;
      p.fill(0,20,30,180);p.stroke(0,180,200,150);p.strokeWeight(1);p.rect(mx,my,mw,mh,5);
      p.noStroke();p.fill(0,220,255);p.textSize(7);p.textAlign(p.LEFT,p.TOP);p.text('DENSITY',mx+4,my+2);
      p.stroke(0,200,220,200);p.strokeWeight(1);p.line(mx,my+8,mx+mw,my+8);
      var dm=[];
      for(var i=0;i<cols;i++){dm[i]=[];for(var j=0;j<rows;j++)dm[i][j]=0;}
      for(var f=0;f<fishes.length;f++){
        if(!fishes[f].dead){
          var cx2=Math.floor(p.map(fishes[f].pos.x,0,CW,0,cols));
          var cy2=Math.floor(p.map(fishes[f].pos.y,0,CH,0,rows));
          cx2=p.constrain(cx2,0,cols-1);cy2=p.constrain(cy2,0,rows-1);dm[cx2][cy2]++;
        }
      }
      p.noStroke();var sy2=my+12;
      for(var i=0;i<cols;i++) for(var j=0;j<rows;j++){
        var d=dm[i][j];
        if(d===0)      p.fill(0,10,20,80);
        else if(d===1) p.fill(0,100,200,150);
        else if(d===2) p.fill(0,200,150,170);
        else if(d===3) p.fill(100,220,50,190);
        else if(d===4) p.fill(220,200,0,210);
        else if(d<8)   p.fill(255,150,0,230);
        else           p.fill(255,50,0,250);
        p.rect(mx+i*gs,sy2+j*gs,gs,gs);
      }
      p.pop();
    }
  }); // end p5
}; // end onload
document.head.appendChild(_sc);

// ── Fish ──────────────────────────────────────────────────────
function Fish(p){
  this.p=p||window._p5i;
  var pp=this.p;
  this.pos=pp.createVector(pp.random(CW),pp.random(CH));
  this.vel=p5.Vector.random2D();this.vel.setMag(pp.random(2,4));
  this.acc=pp.createVector();this.maxForce=0.2;this.size=pp.random(12,20);
  var cc=Math.floor(pp.random(3));
  if(cc===0) this.col=pp.color(pp.random(20,80),pp.random(100,150),pp.random(180,230),200);
  else if(cc===1) this.col=pp.color(pp.random(120,180),pp.random(50,120),pp.random(180,230),200);
  else this.col=pp.color(pp.random(200,255),pp.random(200,255),pp.random(200,255),200);
  this.hunger=pp.random(0,50);this.hungerRate=pp.random(0.02,0.05);
  this.dead=false;this.deathAngle=0;this.opacity=255;this.dv=0;
  this.blinkState=false;this.blinkTimer=0;
}
Fish.prototype.flock=function(fs){
  if(this.dead) return;
  var al=this.align(fs),co=this.cohere(fs),se=this.separate(fs);
  al.mult(P.alignment);co.mult(P.cohesion);se.mult(P.separation);
  this.acc.add(al);this.acc.add(co);this.acc.add(se);
};
Fish.prototype.align=function(fs){
  var p=this.p,st=p.createVector(),tot=0;
  for(var i=0;i<fs.length;i++){
    var d=p.dist(this.pos.x,this.pos.y,fs[i].pos.x,fs[i].pos.y);
    if(fs[i]!==this&&d<P.vision){st.add(fs[i].vel);tot++;}
  }
  if(tot>0){st.div(tot);st.setMag(P.maxSpeed);st.sub(this.vel);st.limit(this.maxForce);}
  return st;
};
Fish.prototype.cohere=function(fs){
  var p=this.p,st=p.createVector(),tot=0;
  for(var i=0;i<fs.length;i++){
    var d=p.dist(this.pos.x,this.pos.y,fs[i].pos.x,fs[i].pos.y);
    if(fs[i]!==this&&d<P.vision){st.add(fs[i].pos);tot++;}
  }
  if(tot>0){st.div(tot);st.sub(this.pos);st.setMag(P.maxSpeed);st.sub(this.vel);st.limit(this.maxForce);}
  return st;
};
Fish.prototype.separate=function(fs){
  var p=this.p,st=p.createVector(),tot=0;
  for(var i=0;i<fs.length;i++){
    var d=p.dist(this.pos.x,this.pos.y,fs[i].pos.x,fs[i].pos.y);
    if(fs[i]!==this&&d<P.vision/2){
      var df=p5.Vector.sub(this.pos,fs[i].pos);df.div(d*d||0.001);st.add(df);tot++;
    }
  }
  if(tot>0){st.div(tot);st.setMag(P.maxSpeed);st.sub(this.vel);st.limit(this.maxForce);}
  return st;
};
Fish.prototype.seekFood=function(pl){
  if(this.dead||this.hunger<30) return;
  var p=this.p,best=null,bd=Infinity;
  for(var i=0;i<pl.length;i++){
    if(pl[i].consumed) continue;
    var d=p.dist(this.pos.x,this.pos.y,pl[i].pos.x,pl[i].pos.y);
    if(d<bd&&d<150){bd=d;best=pl[i];}
  }
  if(best){
    var des=p5.Vector.sub(best.pos,this.pos);des.setMag(P.maxSpeed*1.5);
    var st=p5.Vector.sub(des,this.vel);st.limit(this.maxForce*2);this.acc.add(st.mult(2));
    if(p.dist(this.pos.x,this.pos.y,best.pos.x,best.pos.y)<this.size+best.size){
      best.consumed=true;this.hunger=Math.max(0,this.hunger-20);
    }
  }
};
Fish.prototype.update=function(){
  if(this.dead){
    if(wallOn){
      if(this.pos.y<CH-20){this.dv+=0.15;this.pos.y+=this.dv;}
      else{this.pos.y=CH-20;this.dv=0;this.opacity=Math.max(0,this.opacity-2);}
    } else {this.dv+=0.15;this.pos.y+=this.dv;if(this.pos.y>=CH-20)this.opacity=Math.max(0,this.opacity-8);}
  } else {
    this.pos.add(this.vel);this.vel.add(this.acc);this.vel.limit(P.maxSpeed);this.acc.mult(0);
    if(hungerOn){
      this.hunger=Math.min(100,this.hunger+this.hungerRate);
      if(deathOn&&this.hunger>=85){this.blinkTimer++;if(this.blinkTimer%15===0)this.blinkState=!this.blinkState;}
      else{this.blinkState=false;this.blinkTimer=0;}
      if(deathOn&&this.hunger>=100) this.die();
    } else {this.blinkState=false;this.blinkTimer=0;}
  }
};
Fish.prototype.die=function(){
  this.dead=true;this.deathAngle=this.vel.heading()+this.p.PI;
  this.vel.mult(0);this.col=this.p.color(120,120,120,255);
};
Fish.prototype.edges=function(){
  if(this.dead) return;
  var m=20;
  if(wallOn){
    if(this.pos.x>CW-m){this.vel.x*=-1;this.pos.x=CW-m;}
    else if(this.pos.x<m){this.vel.x*=-1;this.pos.x=m;}
    if(this.pos.y>CH-m){this.vel.y*=-1;this.pos.y=CH-m;}
    else if(this.pos.y<m){this.vel.y*=-1;this.pos.y=m;}
  } else {
    if(this.pos.x>CW+m)this.pos.x=-m;
    else if(this.pos.x<-m)this.pos.x=CW+m;
    if(this.pos.y>CH+m)this.pos.y=-m;
    else if(this.pos.y<-m)this.pos.y=CH+m;
  }
};
Fish.prototype.show=function(){
  var p=this.p,s=this.size;
  p.push();
  var angle=this.dead?this.deathAngle:this.vel.heading();
  p.translate(this.pos.x,this.pos.y);p.rotate(angle);
  var cc=p.color(p.red(this.col),p.green(this.col),p.blue(this.col),this.opacity);
  if(!this.dead&&this.blinkState&&deathOn&&hungerOn) cc=p.color(255,0,0,this.opacity);
  var r=p.red(this.col),g=p.green(this.col),b=p.blue(this.col);
  // Sombra
  p.fill(r-60,g-60,b-60,this.opacity*0.6);p.noStroke();p.ellipse(0,s*0.15,s*1.8,s*0.8);
  // Cuerpo
  p.fill(cc);p.stroke(r-40,g-40,b-40,this.opacity*0.7);p.strokeWeight(1.5);p.ellipse(0,0,s*1.8,s*0.9);
  // Highlight
  p.fill(255,255,255,this.opacity*0.4);p.noStroke();p.ellipse(s*0.2,-s*0.15,s*0.8,s*0.3);
  // Aleta dorsal
  p.fill(r-20,g-20,b-20,this.opacity*0.8);p.stroke(r-60,g-60,b-60,this.opacity*0.8);p.strokeWeight(1);
  p.beginShape();p.vertex(0,-s*0.45);p.vertex(-s*0.3,-s*0.7);p.vertex(-s*0.5,-s*0.5);p.vertex(-s*0.2,-s*0.45);p.endShape(p.CLOSE);
  // Aletas pectorales
  p.fill(r-30,g-30,b-30,this.opacity*0.7);
  p.beginShape();p.vertex(-s*0.2,0);p.vertex(-s*0.5,s*0.6);p.vertex(-s*0.3,s*0.3);p.endShape(p.CLOSE);
  p.beginShape();p.vertex(-s*0.2,0);p.vertex(-s*0.5,-s*0.6);p.vertex(-s*0.3,-s*0.3);p.endShape(p.CLOSE);
  // Cola
  p.fill(r-40,g-40,b-40,this.opacity*0.85);p.stroke(r-70,g-70,b-70,this.opacity*0.8);p.strokeWeight(1.5);
  p.beginShape();p.vertex(-s*0.9,0);p.vertex(-s*1.3,-s*0.5);p.vertex(-s*1.1,-s*0.2);p.endShape(p.CLOSE);
  p.beginShape();p.vertex(-s*0.9,0);p.vertex(-s*1.3,s*0.5);p.vertex(-s*1.1,s*0.2);p.endShape(p.CLOSE);
  // Cabeza
  p.fill(r+20,g+20,b+20,this.opacity);p.noStroke();p.ellipse(s*0.6,0,s*0.8,s*0.7);
  // Ojo
  if(!this.dead){
    p.fill(255,255,255,this.opacity);p.stroke(0,0,0,this.opacity*0.4);p.strokeWeight(1);p.circle(s*0.7,-s*0.1,s*0.25);
    p.fill(0,0,0,this.opacity);p.noStroke();p.circle(s*0.75,-s*0.12,s*0.12);
    p.fill(255,255,255,this.opacity*0.8);p.circle(s*0.78,-s*0.15,s*0.05);
  } else {
    p.stroke(0,0,0,this.opacity*0.6);p.strokeWeight(2);
    var ex=s*0.7,ey=-s*0.1,es=s*0.15;
    p.line(ex-es,ey-es,ex+es,ey+es);p.line(ex-es,ey+es,ex+es,ey-es);
  }
  // Barra hambre
  if(hungerOn&&this.hunger>30&&!this.dead){
    var bw=s*1.5,bh=3,bx=-bw/2,by=-s*1.2;
    p.fill(50,50,50,150);p.noStroke();p.rect(bx,by,bw,bh,2);
    var hp=this.hunger/100;
    p.fill(hp>0.7?p.color(220,20,20):hp>0.4?p.color(220,180,20):p.color(100,220,20));
    p.rect(bx,by,bw*hp,bh,2);
  }
  // Boca
  p.stroke(r-80,g-80,b-80,this.opacity*0.8);p.strokeWeight(1.5);p.noFill();
  p.arc(s*0.85,s*0.05,s*0.15,s*0.1,0,p.PI);
  p.pop();
};

// ── Pellet ────────────────────────────────────────────────────
function Pellet(p,x,y){
  this.p=p;this.pos=p.createVector(x,y);
  this.vel=p.createVector(0,0.5);this.size=p.random(4,7);this.consumed=false;
}
Pellet.prototype.update=function(){
  if(!this.consumed){this.pos.add(this.vel);this.pos.x+=this.p.sin(this.p.frameCount*0.05)*0.3;}
};
Pellet.prototype.show=function(){
  if(!this.consumed){
    this.p.fill(139,69,19);this.p.stroke(89,45,10);this.p.strokeWeight(1);
    this.p.circle(this.pos.x,this.pos.y,this.size);
    this.p.fill(255,200,100,150);this.p.noStroke();
    this.p.circle(this.pos.x-this.size*0.15,this.pos.y-this.size*0.15,this.size*0.4);
  }
};
</script>
</body></html>"""

b64 = base64.b64encode(html.encode('utf-8')).decode('ascii')
IFrame(src="data:text/html;base64," + b64, width=1020, height=640)
```
